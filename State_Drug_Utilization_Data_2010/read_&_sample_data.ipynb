# About 

Below is the process of reading the entrie data formated as .csv and then sample 5% of the entire dataframe. Then it
 writes the sampled dataframe into another csv file. 

## Sources

Links of sources used

- Use "low_memory=False" when loading data (especially large files) and have mixed types Dtype columns.

    - https://github.com/pandas-dev/pandas/issues/5888
    

- Pandas documentation link on how to sample dataframes.
    - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html
    
    
- URL of the original (entire file) "State Drug Utilization Data".
    - https://catalog.data.gov/dataset/state-drug-utilization-data-2010-81ad0/resource/1319364c-c6dc-4ba5-ac9a-53928f710035?inner_span=True

- Or, if .csv format selected.
    - https://data.medicaid.gov/api/views/mmgn-kvy5/rows.csv?accessType=DOWNLOAD

```python
import pandas as pd
```

```python
# Data as .csv format after downloaded from the below URL. 
# https://data.medicaid.gov/api/views/mmgn-kvy5/rows.csv?accessType=DOWNLOAD

df = pd.read_csv("State_Drug_Utilization_Data_2010_original.csv", low_memory=False)
```

```python
df.shape
```

```python
df.head()
```

```python
# generating a sample of the df dataframe. 
df_sample = df.sample(frac =.05, replace=True, random_state=1) 
```

```python
df_sample
```

```python
# Write the reduced dataframe as a .csv file

df_sample.to_csv('State_Drug_Utilization_Data_2010.csv')
```
